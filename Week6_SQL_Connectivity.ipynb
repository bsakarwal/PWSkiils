Q1. What is a database? Differentiate between SQL and NoSQL databases.

A database is a structured collection of data that is organized and stored in a computer system. It allows for efficient data storage, retrieval, manipulation, and management. Databases are commonly used in various applications and systems to store and access data.

SQL and NoSQL are two different types of database management systems with distinct characteristics:

SQL (Structured Query Language) Databases:

   SQL databases are based on the relational model and use SQL as the standard language for interacting with the database.
    They organize data into tables with predefined schemas, where each row represents a record and each column represents a field or attribute.
    SQL databases enforce data integrity through the use of constraints, such as primary keys and foreign keys.
    They provide ACID (Atomicity, Consistency, Isolation, Durability) properties, ensuring data consistency and reliability.
    Examples of SQL databases include MySQL, PostgreSQL, Oracle, Microsoft SQL Server.

NoSQL (Not Only SQL) Databases:

   NoSQL databases are designed to handle large amounts of unstructured or semi-structured data.
    They do not rely on fixed schemas and allow for flexible and dynamic data models.
    NoSQL databases can handle various data types, including key-value pairs, documents, columnar data, and graphs.
    They provide high scalability and performance, making them suitable for handling big data and distributed systems.
    NoSQL databases offer eventual consistency, prioritizing availability and partition tolerance in the CAP theorem.
    Examples of NoSQL databases include MongoDB, Cassandra, Redis, Elasticsearch.

In summary, SQL databases follow a structured approach with predefined schemas and use SQL for querying, while NoSQL databases offer flexibility and scalability for handling diverse data types and volumes without rigid schemas. The choice between SQL and NoSQL depends on the specific requirements of the application, data structure, scalability needs, and desired consistency models.

Q5. Explain Primary Key and Foreign Key.

Primary Key:
    A primary key is a column or set of columns in a table that uniquely identifies each record in the table.
    It ensures that each row in a table is uniquely identified and serves as a unique identifier for that row.
    Primary keys must be unique and cannot contain null values.
    Each table can have only one primary key.
    The primary key constraint is used to define a primary key for a table.

Foreign Key:
    A foreign key is a column or set of columns in one table that refers to the primary key of another table.
    It establishes a relationship between two tables based on the values of the foreign key column(s) and the primary key of the referenced table.
    It ensures referential integrity, meaning that data in the foreign key column(s) must match the values in the referenced primary key column(s).
    Foreign keys can be used to enforce constraints such as enforcing relationships, maintaining data integrity, and enabling table joins.
    Multiple foreign keys can exist in a table, and they can reference different tables or the same table.
    
  Consider two tables, "Orders" and "Customers." The "Orders" table has a foreign key column called "CustomerID," which refers to the primary key column "CustomerID" in the "Customers" table. This establishes a relationship between the two tables, where each order in the "Orders" table is associated with a specific customer in the "Customers" table based on the matching values of the foreign key and primary key columns.
