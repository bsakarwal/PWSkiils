```python
Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.
```

In Flask, the url_for() function is used for URL building. It is a utility function provided by Flask that generates a URL for a given endpoint and any associated arguments.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route("/")
def home():
    return "Welcome to the Home Page!"

@app.route("/about")
def about():
    return "This is the About Page!"

@app.route("/user/<username>")
def user_profile(username):
    return f"Welcome, {username}!"

@app.route("/contact")
def contact():
    return "Contact us at example@example.com"

@app.route("/login")
def login():
    return "Please login to continue"

@app.route("/dashboard")
def dashboard():
    return "Welcome to your dashboard!"

if __name__ == "__main__":
    with app.test_request_context():
        print(url_for("home"))  # Output: /
        print(url_for("about"))  # Output: /about
        print(url_for("user_profile", username="john"))  # Output: /user/john
        print(url_for("contact"))  # Output: /contact
        print(url_for("login"))  # Output: /login
        print(url_for("dashboard"))  # Output: /dashboard
```
