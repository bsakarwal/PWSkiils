Q1. What is MongoDB? Explain non-relational databases in short. In which scenarios it is preferred to use
MongoDB over SQL databases?

MongoDB is a popular open-source NoSQL (non-relational) database management system. It is designed to handle unstructured or semi-structured data, providing high scalability, flexibility, and performance for modern applications.

Non-relational databases, also known as NoSQL databases, are data storage systems that differ from traditional SQL (relational) databases in terms of their data model and querying approach. It is preferred over SQL databases because of below reasons:

Handling Unstructured Data: MongoDB is an excellent choice when dealing with unstructured or semi-structured data like JSON documents. Its flexible document-based model allows for efficient storage, retrieval, and querying of such data.

Scalability and High Throughput: When scalability and high throughput are crucial, MongoDB's ability to distribute data across multiple servers and its efficient write operations make it a suitable choice. It can handle large-scale applications with ease.

Agile Development and Rapid Iteration: MongoDB's flexible schema and ease of schema evolution make it ideal for agile development environments where requirements may change frequently.

Real-time Analytics: MongoDB's support for sharding and its ability to handle large volumes of data make it well-suited for real-time analytics use cases, where quick access to vast amounts of data is essential.

Geospatial Data: MongoDB provides robust geospatial indexing and querying capabilities, making it a preferred choice for applications that deal with location-based data and services.

Q2. State and Explain the features of MongoDB.

1. Flexible Data Model: Non-relational databases do not enforce a fixed schema, allowing for the storage of data in various formats like key-value pairs, documents, graphs, or columnar structures. This flexibility makes them suitable for handling diverse and evolving data types.

2. Horizontal Scalability: Non-relational databases are designed to scale horizontally, meaning they can handle large amounts of data by distributing it across multiple servers or nodes. This enables efficient scaling as data volume or application demands increase.

3. High Performance: Non-relational databases optimize for performance by utilizing data storage models that are closer to the application's data access patterns. They typically employ caching mechanisms and data partitioning to provide fast read and write operations.

4. No Joins: Unlike SQL databases, which often rely on complex joins to retrieve related data from multiple tables, non-relational databases typically avoid joins. Instead, they denormalize data by storing it together, reducing the need for costly join operations.

5. Flexibility in Schema Evolution: Non-relational databases allow for easier schema changes compared to SQL databases. New fields can be added without altering the existing data, making them well-suited for agile development and evolving requirements.

```python
    GET method:
        GET requests are used to retrieve data from a server.
        The data is sent in the URL parameters, appended to the endpoint as key-value pairs.
        GET requests are typically used for data retrieval and should not have any side effects on the server or data.
        GET requests can be cached by browsers and intermediaries, and their parameters are visible in the URL.
        GET requests have limitations on the amount of data that can be sent, usually limited by the maximum length of a URL.
        
    POST method:
        POST requests are used to send data to a server to create or update a resource.
        The data is sent in the request body, separate from the URL.
        POST requests can include more complex data structures, such as JSON or XML, in the request body.
        POST requests are not cached by browsers or intermediaries, and their parameters are not visible in the URL.
        POST requests have no limitations on the amount of data that can be sent.

In summary, the GET method is used for data retrieval, while the POST method is used for creating or updating data on the server. GET requests send data in the URL parameters, while POST requests send data in the request body.
```
