Q2. Write a python program to print Python Exception Hierarchy.

```python
def print_exception_hierarchy(exceptions, indent=0):
    for i in exceptions:
        print(" " * indent + i.__name__)
        if hasattr(i, '__subclasses__'):
            print_exception_hierarchy(i.__subclasses__(), indent + 4)

# Get the base Exception class
base_exception = BaseException

# Get all the built-in exception classes
built_in_exceptions = [i for i in base_exception.__subclasses__() if i.__module__ == 'builtins']

# Print the Exception Hierarchy
print("Python Exception Hierarchy:")
print_exception_hierarchy(built_in_exceptions)
```

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

Q6. List down some best practices for exception handling in python.

1. Do not include super class everytime, call specific error class for identified errors
2. Always print proper messages
3. Always try to log error 
4. Always avoid to write a multiple exception handling
5. Proper Documentation of the code details
6. Clean up the resources eg close a file in finally loop, logging.shutdown after log file open
